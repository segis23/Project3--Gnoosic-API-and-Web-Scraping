```python
soup.select(".chart-element__meta.text--center.color--secondary.text--peak")
```

```python
peak = []
for i in soup.select(".chart-element__meta.text--center.color--secondary.text--peak"):
    peak.append(i.get_text())
```

```python
def recommended_song():
    random_song = random.choice(billboard["song"]) 
    
    user_choice =str(input("Enter your song to get a recommendation: ")).upper()
    
    if user_choice in list(billboard["song"]):
            print("That's a hot song! Keep a try listening to this one:" , random_song, "by", billboard.loc[billboard["song"]== random_song, "artist"].iloc[0])
    else:
        results = spotify.search(q= user_choice)

        song_features = pd.DataFrame(spotify.audio_features(results["tracks"]["items"][0]["uri"]))

        features_selected = song_features[["danceability", "acousticness",
           "energy", "instrumentalness", "liveness", "loudness", "speechiness",
           "tempo"]]


        cluster_assigned = kmeans.predict(scaler.transform(features_selected))
        print(cluster_assigned)

        spoty_song = spotify_df[spotify_df["cluster"] == int(cluster_assigned)][["name","artist"]]
        spoty_return= spoty_song.sample()
        the_song = list(spoty_return["name"])
        the_artist = list(spoty_return["artist"])

        for i,j in zip(the_song, the_artist):
            print("The song Spotify recommends to you is:", i, "by", j)  
            
    return

recommended_song()
```
